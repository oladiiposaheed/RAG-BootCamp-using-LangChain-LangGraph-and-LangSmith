JSON Processing

```python
# Method 1: JsonLoader with jq_schema
emp_loader = JSONLoader(
    file_path='data/json_files/it_company_data.json',
    jq_schema='.employees[]',  # Extract each employee as a separate document
    text_content=False # Do not treat entire JSON as text 
)

emp_docs = emp_loader.load()
print(f"Loaded {len(emp_docs)} documents.")
print(emp_docs[0])  # Print the first document to see its structure
print(f'First employee: {emp_docs[0].page_content[:300]}...')  # Print first 100 characters of content
print(f'First employee skills: {emp_docs[0].metadata.get("skills")}')  # Accessing skills from metadata
print(f'First employee ID: {emp_docs[0].metadata.get("id")}')  # Accessing ID from metadata using get method
# Safely access metadata to avoid KeyError

if hasattr(emp_docs[0], 'metadata'):
    print(emp_docs[0].metadata.get("id", "ID not found"))
else:
    print("No metadata found")
```

```python
# Method 2: Custom Json Processing for complex structures

def process_json(filepath: str) -> List[Document]:
    
    with open(filepath, 'r') as f:
        data = json.load(f)
        
    documents = []
    
    for emp in data.get('employees', []):
        content = f"""Employee Profile:
        Name: {emp['name']}
        Role: {emp['role']}
        Skills: {', '.join(emp['skills'])}
        
        Projects:"""
        for proj in emp.get('projects', []):
            content += f"\n {proj['name']} Status: {proj['status']}"
          
        # Convert to document    
        doc = Document(
            page_content=content,
            metadata={
                'source': filepath,
                'data_type': 'employee_profile',
                'employee_id': emp['id'],
                'employee_name': emp['name'],
                'role': emp['role']
            }
        )
        
        documents.append(doc)
        
    return documents  
```
