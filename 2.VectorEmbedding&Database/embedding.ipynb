### Embedding

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg
```

```python
word_embeddings = {
    'cat': [0.84, 0.72],
    'dog': [0.88, 0.26],
    'kitten': [0.8, 0.8],
    'puppy': [0.9, 0.5],
    'tigger': [0.78, 0.8],
    'truck': [-0.18, -0.3],
    'car': [-0.3, -0.65]
}
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

for word, coords in word_embeddings.items():
    ax.scatter(coords[0], coords[1], s=100)
    ax.annotate(word, (coords[0], coords[1]), xytext=(5, 5), textcoords='offset points')
    
ax.set_xlabel('Dimension 1')
ax.set_ylabel('Dimension 2')
ax.set_title('Word Embeddings Visualization')
ax.grid(True, alpha=0.8)
plt.tight_layout()
plt.show()
    
```

Cosine Similarity

```python
def cosine_similarity(vec_a, vec2_b):
    
    dot_product = np.dot(vec_a, vec2_b)
    norm_vec_a = linalg.norm(vec_a)
    norm_vec_b = linalg.norm(vec2_b)
    
    c_similary = dot_product / (norm_vec_a * norm_vec_b)
    return c_similary
```

```python
# Sample data
cat_vec = [0.8, 0.78, 0.5]
kitten_vec = [0.69, 0.7, 0.43]
dog_vec = [0.2, -0.5, 0.16]
```

```python
cat_kitten_similarity = cosine_similarity(cat_vec, kitten_vec)
print(f'Cat and Kitten Similarity: {cat_kitten_similarity}')
```

```python
cat_dog_similarity = cosine_similarity(cat_vec, dog_vec)
print(f'Cat and dog Similarity: {cat_dog_similarity}')
```
