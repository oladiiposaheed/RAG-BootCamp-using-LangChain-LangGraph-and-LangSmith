### Creating Embeddings with HuggingFace

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings(model_name = 'sentence-transformers/all-MiniLM-L6-v2')
```

```python
embeddings
```

```python
# Create embedding
txt = 'Hi, welcome to our RAG BootCamp Professional course'

import asyncio

embedding = await embeddings.aembed_query(txt)
print(f'Text: {txt}')
print(f'Embedding Length: {len(embedding)}')
print(f'Embedding vector representation:\n {embedding}')
```

```python
# Create embedding
sent = [
    'Hi, welcome to our RAG BootCamp Professional course 1',
    'Hi, welcome to our RAG BootCamp Professional course 2',
    'Hi, welcome to our RAG BootCamp Professional course 3',
    'Hi, welcome to our RAG BootCamp Professional course 4',
    'Hi, welcome to our RAG BootCamp Professional course 5'
]

embedding = embeddings.embed_documents(sent)
print(f'Text: {sent}')
print(f'Number of Embeddings: {len(embedding)}')
print(f'Embedding vector representation:\n {embedding[0]}')
print(f'Embedding vector representation:\n {embedding[1]}')
print(f'Embedding vector representation:\n {embedding[2]}')
print(f'Embedding vector representation:\n {embedding[-1]}')
```
